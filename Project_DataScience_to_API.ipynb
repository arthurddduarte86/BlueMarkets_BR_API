```python
import pandas as pd
```

```python
# leitura dos arquivos
df1 = pd.read_excel('Aracaju.xlsx')
df2 = pd.read_excel('Fortaleza.xlsx')
df3 = pd.read_excel('Natal.xlsx')
df3 = pd.read_excel('Recife.xlsx')
df4 = pd.read_excel('Salvador.xlsx')
```

```python
# juntar/concatenar todos os arquivos em um só
df_global = pd.concat([df1, df2, df3, df4])
```

```python
df_global.sample(10) # 10 amostras aleatórias
```

```python
# tipos de dados das colunas
df_global.dtypes
```

```python
type(df_global)   # é um objeto do tipo dataframe
```

```python
# Verificar se há valores nulos e apresentar o somatório
df_global.isnull().sum()
# se 0, não há
```

```python
df_global.head(0)
# Observe quais colunas o documento possui
```

```python
# Vamos criar uma coluna chamada Receita, que é o faturamento da venda, RECEITA = VENDAS * QUANTIDADE
df_global['Receita'] = df_global['Vendas'].mul(df_global['Qtde'])
df_global.sample(2)
```

```python
# Criando uma nova coluna, recebendo o "ano" de df_global['Data']
df_global['Ano_Venda'] = df_global['Data'].dt.year
# Criando uma nova coluna, recebendo o "mês" de df_global['Data']
df_global['Mes'] = df_global['Data'].dt.month
# Criando uma nova coluna, recebendo o 'dia' de df_global['Data']
df_global['Dia'] = df_global['Data'].dt.day
```

```python
df_global.sample(2)
```

```python
# Criando a coluna de trimestre
df_global['Quarter'] = df_global['Data'].dt.quarter
```

```python
df_global.sample(3)
```

```python
# Padronizando os nomes de PT-BR para EN-US
df_global = df_global.rename(columns={'Cidade': 'City', 
                                      'Data': 'Date', 
                                      'Vendas': 'Value_Price', 
                                      'LojaID': 'ID_City', 
                                      'Qtde': 'TotalProd', 
                                      'Receita': 'Value_Checkout',
                                     'Ano_Venda': 'Year_Checkout', 
                                      'Mes': 'Month_Checkout', 
                                      'Dia': 'Day_Checkout'})
```

```python
df_global.head(2)
```

```python
# Cria uma nova coluna com valores index/indices de contagem e resetando o indice/index
df_global['id_transaction'] = df_global.index
df_global.reset_index(drop=True, inplace=True)
```

```python
df_global.sample(10)
```
